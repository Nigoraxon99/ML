```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

MathScores = np.array([1.17, 2.97, 3.26, 4.69, 5.83, 6.00, 6.41]).reshape(-1, 1)
DrugConcentration = np.array([78.93, 58.20, 67.47, 37.47, 45.65, 32.92, 29.97])

regression = LinearRegression()
regression.fit(DrugConcentration.reshape(-1, 1), MathScores)

slope = regression.coef_[0][0]
intercept = regression.intercept_[0]

plt.scatter(DrugConcentration, MathScores, color='blue', label='Data')
plt.plot(DrugConcentration, regression.predict(DrugConcentration.reshape(-1, 1)), color='red', label='Regression Line')
plt.xlabel('Drug Concentration')
plt.ylabel('Math Scores')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

print(f"Regression Equation: Y = {intercept:.2f} + {slope:.2f} * X")
```
