```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load your data
data = pd.read_csv("R_Projects/tv_station.txt")

# Exclude the "Station Call Letters" column
data = data.drop(columns=["c_Letters"])

# Split the data into X (independent variables) and Y (dependent variable)
X = data.drop(columns=["SalePrice"])
Y = data["SalePrice"]

# Perform multiple linear regression
model = LinearRegression()
model.fit(X, Y)

# Get the coefficients
coefficients = model.coef_
intercept = model.intercept_

# Print the coefficients and intercept
print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
import matplotlib.pyplot as plt
# Predict the values
Y_pred = model.predict(X)

# Create a scatter plot of actual data points
plt.scatter(Y, Y_pred, color="black")

# Create a line plot of the predicted data
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color="blue", linewidth=3)

plt.xlabel("Actual Sale Price")
plt.ylabel("Predicted Sale Price")
plt.title("Actual vs. Predicted Sale Price")

plt.show()
```
